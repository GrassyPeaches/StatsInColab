```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
```

# Normal Distribution
* Unimodal and symmetric, bell shaped curve
* Many variables are nearly normal, but none are exactly normal
* Denoted as N(mew, sigma) --> Normal with  mean (mew) and standard deviation (sigma)





```python
# Pam's SAT DataFrame
df_SAT = pd.DataFrame({'scores': [600, 900, 1200, 1500, 1800, 2100, 2400]})

# Calculate Z-scores
df_SAT['z_score'] = stats.zscore(df_SAT['scores'])

print(df_SAT)
```

```python
# Create x values
x = np.linspace(400, 2400)

# Calculate the PDF(probability density function) of the normal distribution
y = norm.pdf(x, loc=1500, scale=300)

# Plot
plt.plot(x, y, label='Standard Normal', color='purple')
plt.title('Standard Normal Distribution (μ=1500, σ=300)')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

# Here I finally was beginning to understand a little bit more about how to plot the normal curve:

```python
# Pam

mu = 1500 # normal mean of SAT scores
sigma = 300 # standard deviation
pam_score = 1800 # Pam's SAT score

x = np.linspace(600, 2400, 1000) # x- values: SAT score ranges

y = norm.pdf(x, mu, sigma) # y values that define bell curve

pam_zscore = (pam_score -mu) / sigma # z-score formula
pam_prob_density = norm.pdf(pam_score, mu, sigma)

plt.plot(x, y, label='Normal Distribution', color='blue')
plt.axvline(pam_score, color='red', linestyle='dashed', label='Pam\'s Score (1800)')

plt.title('Normal Distribution of SAT Scores\n($\mu$=1500, $\sigma$=300)')
plt.xlabel('SAT Score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

print(f"Pam's Z-Score is {pam_zscore:.2f}, which means it is within 1 standard deviation of the mean.")
```

```python
# Jim

mu = 21 # normal mean of ACT scores
sigma = 5 # standard deviation
jim_score = 24 # Jim's ACT score

x = np.linspace(6, 36, 1000) # x- values: SAT score ranges

y = norm.pdf(x, mu, sigma) # y values that define bell curve

jim_zscore = (jim_score -mu) / sigma # z-score formula
jim_prob_density = norm.pdf(pam_score, mu, sigma)

plt.plot(x, y, label='Normal Distribution', color='red')
plt.axvline(jim_score, color='blue', linestyle='dashed', label='Jim\'s Score (24)')

plt.title('Normal Distribution of ACT Scores\n($\mu$=21, $\sigma$=5)')
plt.xlabel('ACT Score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

print(f"Jim's Z-Score is {jim_zscore:.2f}, which means it is within .6 of the standard deviation of the mean.")
```
